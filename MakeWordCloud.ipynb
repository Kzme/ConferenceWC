```python
import requests
from bs4 import BeautifulSoup
from wordcloud import WordCloud

url = 'https://papers.nips.cc/book/advances-in-neural-information-processing-systems-31-2018'
r = requests.get(url)
soup = BeautifulSoup(r.text, "html.parser")

# タイトルを探して表示
title_stock = ""
span = soup.find_all("a")
for s in span:
    if "/paper/" in s.get("href"):
        title = s.string
        if title is not None:
            title_stock = title_stock + s.string + '  '
#title_stock

wordcloud = WordCloud(
    font_path="/Library/Fonts/AppleGothic.ttf",      #使用フォント
    scale=2.0,                                                                   #出力画像のサイズ
    background_color="white",                                     #背景色
    min_font_size=10,                                                   #最小フォントサイズ
    max_font_size=100,                                                  #最大フォントサイズ
   prefer_horizontal=90,                                                #文字列回転角
    stopwords= [
                "via",
                "for",
                "and",
                "with",
                "from",
                "to",
                "in",
                "of",
                "the",
                "on",
                "Learning",
                "Network",
                "Deep",
                "Networks",
                "Data",
                "using",
                "An",
                "Model",
                "Models",
                "Neural",
                "through",
                "by",
                "Algorithm",
                "Efficient",
                "Non"],
    width=800,height=600).generate(title_stock)
wordcloud.to_file("./wordcloud_test.png")
```

```python
def MakeWordCloud(title_stock, outname):
    wordcloud = WordCloud(
        font_path="/Library/Fonts/AppleGothic.ttf",      #使用フォント
        scale=2.0,                                                                   #出力画像のサイズ
        background_color="white",                                     #背景色
        min_font_size=10,                                                   #最小フォントサイズ
        max_font_size=100,                                                  #最大フォントサイズ
       prefer_horizontal=90,                                                #文字列回転角
        stopwords= [
                    "via",
                    "for",
                    "and",
                    "with",
                    "from",
                    "to",
                    "in",
                    "of",
                    "the",
                    "on",
                    "Learning",
                    "Network",
                    "Deep",
                    "Networks",
                    "Data",
                    "using",
                    "An",
                    "Model",
                    "Models",
                    "Neural",
                    "through",
                    "by",
                    "Algorithm",
                    "Efficient",
                    "Non",
                    "Optimization",
                    "based",
                    "Method"],
        width=800,height=600).generate(title_stock)
    wordcloud.to_file(outname)
```

```python
import requests
from bs4 import BeautifulSoup
from wordcloud import WordCloud

url = 'https://papers.nips.cc/'
r = requests.get(url)
soup = BeautifulSoup(r.text, "html.parser")
ExtractedTag  = soup.find_all("a")
```

```python
for tmpString in ExtractedTag:
    #print(tmpString.get("href"))
    if "/book/" in tmpString.get("href"):
        url = "".join(["https://papers.nips.cc", tmpString.get("href")])
        #print(url)
        Year = url.split('-')[-1]
        OutName = "".join([Year, "NIPS", "_WordCloud.png"])
        print(OutName)
        r = requests.get(url)
        TitleSoup = BeautifulSoup(r.text, "html.parser")
        # タイトルを探して表示
        title_stock = ""
        span = TitleSoup.find_all("a")
        for s in span:
            if "/paper/" in s.get("href"):
                title = s.string
                if title is not None:
                    title_stock = title_stock + s.string + '  '
        #print(title_stock)
        MakeWordCloud(title_stock, OutName)
```
